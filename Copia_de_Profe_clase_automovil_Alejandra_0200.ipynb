```python
class Automovil:
  def __init__(self, marca, placa, precio):
    self.marca = marca
    self.placa = placa
    self.precio = precio

  def leerdatos(self):
    self.marca = input("ingrese la marca: ")
    self.placa = input("ingrese la placa:")
    # Convert price to a numeric type if intended, otherwise keep as string
    self.precio = input("ingrese el precio: ")

  def mostrardatos(self):
    print("marca:", self.marca)
    print("placa:", self.placa)
    print("precio:", self.precio)
# fin de la clase automovil

print("<3-<3-<3-<3-<3-<3-<3")

# Mostrando datos para cada carro
print("--- Datos del Carro 1 ---")
carro1 = Automovil("", "", "")
carro1.mostrardatos()
carro1.leerdatos()
print("<3-<3-<3-<3-<3-<3-<3")
print("--- Datos del Carro 2 ---")
carro2 = Automovil("", "", "")
carro2.mostrardatos()
carro2.leerdatos()
print("<3-<3-<3-<3-<3-<3-<3")
print("--- Datos del Carro 3 ---")
carro3 = Automovil("", "", "")
carro3.mostrardatos()
carro3.leerdatos()
print("<3-<3-<3-<3-<3-<3-<3")
print("sofware desarollado por Alejandra Rivera Valenzuela NL=45")
```
